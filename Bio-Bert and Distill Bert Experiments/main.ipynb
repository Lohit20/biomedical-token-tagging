```python
from google.colab import drive
import os
import nbformat
from nbconvert.preprocessors import ExecutePreprocessor
from google.colab import userdata
import wandb
```

```python
my_secret_key = userdata.get('HF_TOKEN')
wb_token = userdata.get("wandb")

wandb.login(key=wb_token)
run = wandb.init(
    project='Fine-tuning bert series for Biomedical Text annotation',
    job_type="training",
    anonymous="allow"
)
```

```python
drive.mount('/content/drive')
notebook_folder = '/content/drive/MyDrive/NLP_Course_work'
```

```python
ep = ExecutePreprocessor(timeout=600, kernel_name='python3')

for notebook_name in os.listdir(notebook_folder):
    if notebook_name.endswith('.ipynb') and notebook_name != 'main.ipynb':
        notebook_path = os.path.join(notebook_folder, notebook_name)
        print(f'\n Running notebook: {notebook_name}...\n{"="*60}')

        with open(notebook_path) as f:
            nb = nbformat.read(f, as_version=4)

        try:
            ep.preprocess(nb, {'metadata': {'path': notebook_folder}})
            for cell in nb.cells:
                if 'outputs' in cell:
                    for output in cell['outputs']:
                        if output.output_type == 'stream':
                            print(output.text)
                        elif output.output_type == 'execute_result' and 'text/plain' in output.data:
                            print(output.data['text/plain'])
                        elif output.output_type == 'error':
                            print('ERROR:', '\n'.join(output['traceback']))
            print(f'\n {notebook_name} completed successfully.\n{"="*60}')
        except Exception as e:
            print(f'\n Error running {notebook_name}:\n{e}\n{"="*60}')
```
